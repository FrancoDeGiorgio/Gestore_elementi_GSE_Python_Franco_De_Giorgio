```python
# --- SETUP INIZIALE ---

def genera_id():
    """Genera un ID univoco incrementale per un nuovo capo.
    
    Returns:
        int: Il prossimo ID disponibile, basato sul massimo ID esistente + 1.
    """
    return max([capo["id"] for capo in capi_lista], default=0) + 1



# --- FUNZIONI BASE ---

def insert_capo():
    """Inserisce un nuovo capo nell'inventario tramite input utente.
    
    Guida l'utente attraverso l'inserimento di tutte le informazioni necessarie
    per un nuovo capo d'abbigliamento: nome, tipologia, taglia, colore, marca,
    condizione, disponibilit√† e prezzo (se in vendita).
    
    Returns:
        None
    """
    print("\n--- Inserisci un nuovo capo ---")
    nome = input("Inserisci il nome del nuovo prodotto: ").strip()

    # Selezione della tipologia del capo
    print("\nScegli la tipologia :")
    for key, value in tipi.items():
        print(f"{key}. {value}")

    # Ciclo per assicurarsi che l'utente selezioni una tipologia valida
    while True:
        tipo_input = input("Scegli fra queste tipologie (1-9): ")
        if tipo_input in tipi:
            tipo = tipi[tipo_input]
            break
        else:
            print("Opzione non valida. Riprova.")

    # Selezione della taglia
    print("\nScegli la taglia:")
    for key, value in taglie.items():
        print(f"{key}. {value}")

    # Ciclo per assicurarsi che l'utente selezioni una taglia valida
    while True:
        taglia_input = input("Seleziona la taglia (1-9): ")
        if taglia_input in taglie:
            taglia = taglie[taglia_input]
            break
        else:
            print("Opzione non valida. Riprova.")

    # Selezione del colore
    print("\nScegli il colore:")
    for key, value in colori.items():
        print(f"{key}. {value}")

    # Ciclo per assicurarsi che l'utente selezioni un colore valido
    while True:
        colore_input = input("Seleziona il colore (1-11): ")
        if colore_input in colori:
            colore = colori[colore_input]
            break
        else:
            print("Opzione non valida. Riprova.")

    # Inserimento della marca
    marca = input("Inserisci la marca del nuovo prodotto: ").strip()

    # Selezione delle condizioni del capo
    print("\nScegli la condizione:")
    for key, value in condizione.items():
        print(f"{key}. {value}")

    # Ciclo per assicurarsi che l'utente selezioni una condizione valida
    while True:
        condizione_input = input("Seleziona la condizione (1-4): " )
        if condizione_input in condizione:
            condizioni = condizione[condizione_input]
            break
        else:
            print("Opzione non valida. Riprova.")


   # Selezione della disponibilit√† con integrata funzione di stima prezzo
    print("\nModalit√† di disponibilit√†:")
    for key, value in disponibilita.items():
        print(f"{key}. {value}")

    # Ciclo per assicurarsi che l'utente selezioni una disponibilit√† valida
    # e gestisca correttamente l'inserimento del prezzo se necessario
    while True:
        disp_input = input("Seleziona la disponibilit√† (1-3): ")
        if disp_input in disponibilita:
            disposizione = disponibilita[disp_input]
            
            # Se la disponibilit√† √® "vendita", si gestisce il prezzo
            if disp_input == "1":  # Vendita
                while True:
                    # Proposta del sistema di stima automatica del prezzo
                    scelta_stima = input("Vuoi che il sistema ti consigli un prezzo? (s/n): ").strip().lower()
                    if scelta_stima == "s":
                         # Calcolo del prezzo stimato basato su categoria e condizioni
                        prezzo_stimato = stima_prezzo(tipo, condizioni)
                        print(f"üí° Il sistema suggerisce un prezzo di: ‚Ç¨{prezzo_stimato:.2f}")
                        # Conferma dell'utilizzo del prezzo stimato
                        while True:
                            conferma = input("Vuoi usare questo prezzo ? (s/n): ").strip().lower()
                            if conferma == "s":
                                prezzo = prezzo_stimato
                                break  # esci da scelta_stima
                            elif conferma == "n":
                                 # Inserimento manuale del prezzo con validazione
                                while True:                    
                                    try:
                                        prezzo = float(input("Inserisci il prezzo: "))
                                        if prezzo < 0:
                                            raise ValueError
                                        break
                                    except ValueError:
                                        print("‚ö†Ô∏è Prezzo non valido. Riprova.")
                                break                        
                            else:
                                print("‚ö†Ô∏è Inserisci solo 's' per s√¨ o 'n' per no.")
                        break  # esci da scelta_stima
                    elif scelta_stima == "n":
                        while True:
                            try:
                                prezzo = float(input("Inserisci il prezzo: "))
                                if prezzo < 0:
                                    raise ValueError
                                break
                            except ValueError:
                                print("‚ö†Ô∏è Prezzo non valido. Riprova.")
                        break
                    else:
                        print("‚ö†Ô∏è Inserisci solo 's' per s√¨ o 'n' per no.")
            else:
                # Se non √® in vendita, il prezzo √® impostato a 0
                prezzo = 0.0
            break
        else:
            print("‚ö†Ô∏è Opzione non valida. Riprova.")    

     # Creazione e salvataggio del capo nella lista
    capo = {
        "id": genera_id(),
        "nome": nome,
        "categoria": tipo,
        "taglia": taglia,
        "colore": colore,
        "marca": marca,
        "modalit√†_disponibilit√†": disposizione,
        "prezzo": prezzo,
        "condizione": condizioni,
        
    }

    # Aggiunta del capo alla lista e conferma
    capi_lista.append(capo)
    print(f"\n‚úÖ {nome} √® stato aggiunto con successo a capi_lista!\n")
    print("\nüßæ Riepilogo capo aggiunto:")
    for k, v in capo.items():
        print(f" - {k}: {v}")
    


def stampa_tabella_capi(lista, messaggio=None, nessun_risultato="Nessun capo trovato."):
    """Stampa una tabella formattata dei capi d'abbigliamento.
    
    Args:
        lista (list): Lista di dizionari contenente i capi da visualizzare.
        messaggio (str, optional): Messaggio da mostrare prima della tabella.
        nessun_risultato (str, optional): Messaggio da mostrare se la lista √® vuota.
            Default: "Nessun capo trovato."
    
    Returns:
        None
    """
        if lista:
            if messaggio:
                print(messaggio)
                 # Intestazione della tabella
            print(f"{'ID':<3} {'Nome':<20} {'Marca':<15} {'Taglia':<6} {'Colore':<15} {'Prezzo (‚Ç¨)':<11} {'Condizione':<30} {'Disponibilit√†'}")
            print("-" * 120)
            # Righe della tabella
            for capo in lista:
                print(f"{capo['id']:<3} {capo['nome']:<20} {capo['marca']:<15} {capo['taglia']:<6} {capo['colore']:<15} "
                      f"{capo['prezzo']:<11.2f} {capo['condizione']:<30} {capo['modalit√†_disponibilit√†'].capitalize()}")
        else:
            print(nessun_risultato)   



def view_capi():
     """Visualizza tutti i capi disponibili nell'inventario.
    
    Mostra tutti i capi in formato tabella e offre la possibilit√† di 
    aggiungere articoli al carrello.
    
    Returns:
        None
    """
    if not capi_lista:
        print("‚ùå Nessun capo disponibile al momento.")
        return
        
    # Visualizza i capi in formato tabella
    stampa_tabella_capi(capi_lista, messaggio="üì¶ Prodotti disponibili:\n")
    # Offre la possibilit√† di aggiungere capi al carrello
    aggiungi_al_carrello(capi_lista)



# Permette la modifica di un capo esistente nell'inventario.
# L'utente pu√≤ modificare tutti i dettagli del capo selezionato tramite ID,
# con la possibilit√† di mantenere i valori esistenti premendo invio.
def modifica_capo():
    """Modifica un capo esistente nell'inventario.
    
    Permette all'utente di modificare tutti i dettagli del capo selezionato
    tramite ID, con la possibilit√† di mantenere i valori esistenti premendo invio.
    
    Returns:
        None
    """
    print("\n--- Modifica un capo ---")
    try:
        src_id = int(input("üîç Inserisci l'ID del prodotto da modificare: "))
    except ValueError:
        print("‚ö†Ô∏è L'ID deve essere un numero.")
        return

    # Cerca il capo con l'ID specificato
    for capo in capi_lista:
        if capo["id"] == src_id:
            print(f"Trovato capo: {capo['nome']} (ID: {src_id})")

            # Modifica propriet√†: per ognuna si pu√≤ mantenere il valore attuale o inserirne uno nuovo
            nuovo_nome = input(f"Inserisci il nuovo nome [{capo['nome']}]: (Premi Invio per mantenere il valore corrente) ") or capo["nome"]
            nuova_marca = input(f"Inserisci la nuova marca [{capo['marca']}]: (Premi Invio per mantenere il valore corrente) ") or capo["marca"]

            # Modifica categoria
            print("Scegli la nuova categoria:")
            for k, v in tipi.items():
                print(f"{k}. {v}")
            nuova_categoria = tipi.get(input(f"Scelta [{capo['categoria']}]: (Premi Invio per mantenere il valore corrente) "), capo["categoria"])

            # Modifica taglia
            print("Scegli la nuova taglia:")
            for k, v in taglie.items():
                print(f"{k}. {v}")
            nuova_taglia = taglie.get(input(f"Scelta [{capo['taglia']}]: (Premi Invio per mantenere il valore corrente) "), capo["taglia"])

            # Modifica colore
            print("Scegli il nuovo colore:")
            for k, v in colori.items():
                print(f"{k}. {v}")
            nuovo_colore = colori.get(input(f"Scelta [{capo['colore']}]: (Premi Invio per mantenere il valore corrente) "), capo["colore"])

            # Modifica prezzo con gestione errori
            try:
                nuovo_prezzo = input(f"Inserisci il nuovo prezzo [{capo['prezzo']} ‚Ç¨]: ")
                nuovo_prezzo = float(nuovo_prezzo) if nuovo_prezzo else capo["prezzo"]
            except ValueError:
                print("‚ö†Ô∏è Prezzo non valido. Mantenuto il valore precedente.")
                nuovo_prezzo = capo["prezzo"]

            # Modifica condizione
            print("Scegli la nuova condizione:")
            for k, v in condizione.items():
                print(f"{k}. {v}")
            nuova_condizione = condizione.get(input(f"Scelta [{capo['condizione']}]: (Premi Invio per mantenere il valore corrente) "), capo["condizione"])
            
            print("Scegli la nuova modalit√† di disponibilit√†:")
            for k, v in disponibilita.items():
                print(f"{k}. {v}")
            nuova_disponibilita = disponibilita.get(input(f"Scelta [{capo['modalit√†_disponibilit√†']}]: (Premi Invio per mantenere il valore corrente) "), capo["modalit√†_disponibilit√†"])

            # Conferma delle modifiche
            conferma = input(f"\nConfermi le modifiche per il capo ID {src_id}? (s/n): ").strip().lower()
            if conferma == 's':
                # Aggiorna tutti i valori del capo con quelli nuovi
                capo.update({
                    "nome": nuovo_nome,
                    "marca": nuova_marca,
                    "categoria": nuova_categoria,
                    "taglia": nuova_taglia,
                    "colore": nuovo_colore,
                    "prezzo": nuovo_prezzo,
                    "condizione": nuova_condizione,
                    "modalit√†_disponibilit√†": nuova_disponibilita
                })

                print("‚úÖ Capo aggiornato con successo.")
            else:
                print("‚ùå Modifica annullata.")
            return
     #l'ID non √® stato trovato
    print(f"‚ùå Nessun capo trovato con ID {src_id}.")


def elimina_capo():
    """Elimina uno o pi√π capi dall'inventario.
    
    Permette l'eliminazione di uno o pi√π capi esistenti nella lista,
    selezionandoli tramite ID. Richiede conferma prima di ogni eliminazione.
    
    Returns:
        None
    """
    print("\n--- Elimina capo/i ---")
    
    # Richiede all'utente uno o pi√π ID separati da virgola
    input_ids = input("Inserisci uno o pi√π ID da eliminare (separati da virgola): ").strip()
    
    # Se l'utente preme solo invio, interrompe la funzione
    if not input_ids:
        print("‚ö†Ô∏è Nessun ID inserito.")
        return

    try:
        # Converte gli ID inseriti in una lista di interi, rimuovendo spazi bianchi
        id_list = [int(i.strip()) for i in input_ids.split(',')]
    except ValueError:
        # Se uno degli ID non √® numerico, segnala errore e termina
        print("‚ö†Ô∏è Tutti gli ID devono essere numerici.")
        return

    eliminati = 0 
    
    for src_id in id_list:
        # Cerca il capo nella lista tramite ID 
        capo = next((c for c in capi_lista if c["id"] == src_id), None)
        
        if capo:
            # Chiede conferma prima di eliminare
            conferma = input(f"Eliminare '{capo['nome']}' (ID {src_id})? (s/n): ").strip().lower()
            if conferma == "s":
                capi_lista.remove(capo)  # Rimuove il capo dalla lista
                eliminati += 1
                print(f"‚úÖ Capo ID {src_id} eliminato.")
            else:
                print(f"‚ùé Capo ID {src_id} non eliminato.")
        else:
            # ID non trovato nella lista
            print(f"‚ùå Nessun capo trovato con ID {src_id}.")

    # Riepilogo finale
    if eliminati:
        print(f"\n‚úÖ Totale capi eliminati: {eliminati}")
    else:
        print("\n‚ÑπÔ∏è Nessun capo eliminato.")



def ricerca_capo():
    """Ricerca capi nell'inventario in base a criteri testuali.
    
    Permette la ricerca di capi in base a diversi campi (nome, marca, categoria,
    taglia, colore, condizione) e mostra i risultati corrispondenti. Offre anche
    la possibilit√† di aggiungere articoli trovati al carrello.
    
    Returns:
        None
    """
    print("\n--- Ricerca Capi ---")
    print("Puoi cercare per nome, marca, categoria, taglia, colore o condizione.")
    criterio = input("Inserisci un criterio di ricerca: ").strip().lower()

    # Lista per i risultati della ricerca
    risultati = []

    # Cerca il criterio in vari campi di ogni capo (ricerca case-insensitive)
    for capo in capi_lista:
        if(criterio in capo["nome"].lower() or
           criterio in capo["marca"].lower() or
           criterio in capo["categoria"].lower() or
           criterio in capo["taglia"].lower() or
           criterio in capo["colore"].lower() or
           criterio in capo["condizione"].lower()):
           risultati.append(capo)

    # Visualizza i risultati
    stampa_tabella_capi(
        risultati, 
        messaggio=f"\nTrovati {len(risultati)} capi corrispondenti:\n", 
        nessun_risultato="Nessun capo trovato con il criterio inserito."
    )

    # Se ci sono risultati, offre l'opzione di aggiungere al carrello
    if risultati:
        aggiungi_al_carrello(risultati)


# --- FUNZIONI AVANZATE ---

def filtra_capi():
    """Filtra i capi usando criteri multipli avanzati.
    
    Permette all'utente di applicare diversi filtri contemporaneamente
    (categoria, taglia, colore, condizione, disponibilit√†, prezzo) per trovare
    capi che soddisfano criteri specifici.
    
    Returns:
        None
    """
    print("\n--- Filtra Capi Avanzato ---")
    print("Puoi filtrare i capi per pi√π criteri. Lascia vuoto un filtro se non vuoi applicarlo.\n")

    # Filtro per categoria
    print("Categoria:")
    for k, v in tipi.items():
        print(f"{k}: {v}")
    scelta_categoria = input("Seleziona una categoria (invio per saltare): ").strip()
    categoria = tipi.get(scelta_categoria, "") if scelta_categoria else ""

    # Filtro per taglia
    print("\nTaglia:")
    for k, v in taglie.items():
        print(f"{k}: {v}")
    scelta_taglia = input("Seleziona una taglia (invio per saltare): ").strip()
    taglia = taglie.get(scelta_taglia, "") if scelta_taglia else ""

    # Filtro per colore
    print("\nColore:")
    for k, v in colori.items():
        print(f"{k}: {v}")
    scelta_colore = input("Seleziona un colore (invio per saltare): ").strip()
    colore = colori.get(scelta_colore, "") if scelta_colore else ""

    # Filtro per condizione
    print("\nCondizione:")
    for k, v in condizione.items():
        print(f"{k}: {v}")
    scelta_condizione = input("Seleziona una condizione (invio per saltare): ").strip()
    condizione_sel = condizione.get(scelta_condizione, "") if scelta_condizione else ""

    # Filtro per disponibilit√†
    print("\nModalit√† di disponibilit√†:")
    for k, v in disponibilita.items():
        print(f"{k}: {v}")
    scelta_disponibilita = input("Seleziona la modalit√† (invio per saltare): ").strip()
    disponibilita_sel = disponibilita.get(scelta_disponibilita, "") if scelta_disponibilita else ""

    # Filtro per prezzo
    prezzo_min = input("\nPrezzo minimo (‚Ç¨) (invio per saltare): ").strip()
    prezzo_max = input("Prezzo massimo (‚Ç¨) (invio per saltare): ").strip()

    # Applicazione dei filtri
    risultati = []
    for capo in capi_lista:
        # Verifica tutti i filtri testuali
        if (
            (not categoria or capo["categoria"] == categoria) and
            (not taglia or capo["taglia"] == taglia) and
            (not colore or capo["colore"] == colore) and
            (not condizione_sel or capo["condizione"].lower() == condizione_sel.lower()) and
            (not disponibilita_sel or capo["modalit√†_disponibilit√†"].lower() == disponibilita_sel.lower())
        ):
            # Verifica i filtri numerici (prezzo
            prezzo = capo["prezzo"]
            if (
                (not prezzo_min or prezzo >= float(prezzo_min)) and
                (not prezzo_max or prezzo <= float(prezzo_max))
            ):
                risultati.append(capo)

    # Visualizza i risultati e offre l'opzione di aggiungere al carrello
    stampa_tabella_capi(
        risultati, 
        messaggio=f"\nTrovati {len(risultati)} capi corrispondenti:\n", 
        nessun_risultato="Nessun capo trovato con i filtri inseriti.")
    if risultati:
        aggiungi_al_carrello(risultati)



def stima_prezzo(categoria, condizione):
    """Calcola un prezzo stimato per un capo.
    
    Args:
        categoria (str): La categoria del capo.
        condizione (str): La condizione del capo.
    
    Returns:
        float: Il prezzo stimato basato su categoria e condizione.
    """
    # Ottiene il prezzo base dalla categoria o usa 20 come valore predefinito
    base = prezzo_base_categoria.get(categoria, 20)
    # Ottiene il moltiplicatore dalla condizione o usa 0.5 come valore predefinito
    moltiplicatore = moltiplicatore_condizione.get(condizione, 0.5)

    # Calcola il prezzo stimato
    return base * moltiplicatore



def statistiche_capi():
     """Genera statistiche e grafici sull'inventario attuale.
    
    Utilizza matplotlib per creare grafici che mostrano la distribuzione
    dei capi per categoria, condizione e disponibilit√†.
    
    Returns:
        None
    """
     
    # Conta le categorie nei capi
    for capo in capi_lista:
        categoria = capo["categoria"]
        condizione = capo["condizione"]
        disponibilit√† = capo["modalit√†_disponibilit√†"]
        # Incrementa i contatori appropriati
        if categoria in conteggio_categorie:
            conteggio_categorie[categoria] += 1
        if condizione in conteggio_condizione:
            conteggio_condizione[condizione] += 1
        if disponibilit√† in conteggio_disponibilit√†:
            conteggio_disponibilit√†[disponibilit√†] += 1
            
    # Estrai le chiavi e i valori per il grafico delle categorie
    x = list(conteggio_categorie.keys())
    y = list(conteggio_categorie.values())

    # Imposta la figura complessiva
    plt.figure(figsize=(15,5))

    # PRIMO GRAFICO: Distribuzione per categoria (grafico a barre)
    plt.subplot(1, 3, 1)    
    plt.bar(x, y, color='orange')

    # Aggiungi i numeri sopra le barre
    for i in range(len(x)):
        plt.text(x[i], y[i] + 0.1, str(y[i]), ha='center', va='bottom')  # Aggiungi i numeri sopra le barre

    # Imposta l'orientamento delle etichette sull'asse x
    plt.xticks(rotation=45)

    # Aggiungi titolo e etichette
    plt.title("Quantit√† per categorie")
    plt.xlabel("Categoria")
    plt.ylabel("Quantit√†")
    
    # SECONDO GRAFICO: Distribuzione per condizione (grafico a torta)
    plt.subplot(1, 3, 2)

    x_pie1 = list(conteggio_condizione.keys())
    y_pie1 = list(conteggio_condizione.values())

    plt.pie(y_pie1, labels=x_pie1, autopct='%1.1f%%')
    plt.title("Percentuale capi per condizione")

   # TERZO GRAFICO: Distribuzione per disponibilit√† (grafico a torta)
    plt.subplot(1, 3, 3)

    x_pie2 = list(conteggio_disponibilit√†.keys())
    y_pie2 = list(conteggio_disponibilit√†.values())

    plt.pie(y_pie2, labels=x_pie2, autopct='%1.1f%%')
    plt.title("Percentuale capi per disponibilit√†")

    # Rende il layout pi√π compatto
    plt.tight_layout()
    # Mostra i grafici
    plt.show()



def view_carrello():
    """Visualizza il contenuto attuale del carrello e il totale.
    
    Mostra tutti i capi presenti nel carrello in formato tabella
    e calcola il totale del prezzo.
    
    Returns:
        None
    """
    print("\nüõí Riepilogo carrello:")
    if not carrello:
        print("Il carrello √® vuoto.")
        return

    # Visualizza i capi nel carrello
    stampa_tabella_capi(carrello, messaggio="", nessun_risultato="")
    # Calcola e mostra il totale
    totale = sum(capo["prezzo"] for capo in carrello)
    print(f"\nTotale: {totale:.2f} ‚Ç¨")



def menu_carrello():
    """Gestisce il menu del carrello con varie opzioni.
    
    Presenta un menu interattivo per la gestione del carrello con opzioni
    per visualizzare, rimuovere articoli, svuotare il carrello o tornare
    al menu principale.
    
    Returns:
        None
    """
    while True:
        print("\n--- Gestione Carrello ---")
        print("1. Visualizza contenuto carrello")
        print("2. Rimuovi un capo dal carrello")
        print("3. Svuota carrello")
        print("4. Torna al menu principale")

        scelta = input("Scegli un'opzione: ").strip()

        if scelta == "1":
            view_carrello()
        elif scelta == "2":
            rimuovi_dal_carrello()
        elif scelta == "3":
            svuota_carrello()
        elif scelta == "4":
            break
        else:
            print("‚ùå Scelta non valida.")



def aggiungi_al_carrello(lista_capi):
    """Permette di aggiungere capi al carrello.
    
    Chiede all'utente gli ID dei capi da aggiungere e verifica che
    l'ID esista nella lista e che non sia gi√† nel carrello.
    
    Args:
        lista_capi (list): Lista dei capi da cui selezionare per l'aggiunta.
    
    Returns:
        None
    """
    
    while True:
        scelta = input("Vuoi aggiungere uno o pi√π capi al carrello? (s/n): ").strip().lower()
        if scelta == "n":
            return
        elif scelta == "s":
            # Chiede gli ID dei capi da aggiungere
            ids = input("Inserisci gli ID dei capi separati da virgola (es. 1,3,5): ").split(",")
            aggiunti = 0
            for id_input in ids:
                id_input = id_input.strip()
                if not id_input.isdigit():
                    print(f"‚ö†Ô∏è  ID non valido: '{id_input}'")
                    continue
                    
                id_capo = int(id_input)
                # Cerca il capo nella lista
                capo = next((c for c in lista_capi if c["id"] == id_capo), None)
                
                if not capo:
                    print(f"‚ùå Nessun capo trovato con ID {id_capo}")
                elif any(c["id"] == id_capo for c in carrello):
                    print(f"‚ö†Ô∏è  Capo con ID {id_capo} √® gi√† nel carrello.")
                else:
                    # Aggiunge il capo al carrello
                    carrello.append(capo)
                    print(f"‚úÖ Capo con ID {id_capo} aggiunto al carrello.")
                    aggiunti += 1
                    
            if aggiunti > 0:
                print(f"\nüîÑ Totale articoli nel carrello: {len(carrello)}")
            return
        else:
            print("Inserisci 's' per s√¨ o 'n' per no.")



def rimuovi_dal_carrello():
    """Permette all'utente di rimuovere capi dal carrello.
    
    Mostra il contenuto attuale del carrello e permette all'utente
    di selezionare un capo da rimuovere tramite ID. Continua fino
    a quando l'utente preme INVIO per uscire.

    Returns:
        None
    """
    
    if not carrello:
        print("üõí Il carrello √® vuoto, nulla da rimuovere.")
        return
         
    # Mostra il contenuto attuale del carrello
    view_carrello()
    while True:
        scelta = input("Inserisci l'ID del capo da rimuovere (INVIO per uscire): ").strip()
        if scelta == "":
            break
            
        if not scelta.isdigit():
            print("‚ö†Ô∏è Inserisci un ID numerico valido.")
            continue
            
        id_da_rimuovere = int(scelta)
        # Cerca il capo nel carrello
        capo_da_rimuovere = next((c for c in carrello if c['id'] == id_da_rimuovere), None)
        
        if capo_da_rimuovere:
            # Rimuove il capo dal carrello
            carrello.remove(capo_da_rimuovere)
            print(f"‚ùå Capo '{capo_da_rimuovere['nome']}' rimosso dal carrello.")
        else:
            print("‚ö†Ô∏è ID non trovato nel carrello.")



def svuota_carrello():
    """Svuota completamente il carrello dopo conferma dell'utente.
    
    Chiede conferma all'utente prima di procedere con lo svuotamento.
    Se il carrello √® gi√† vuoto, informa l'utente.

    Returns:
        None
    """
    
    global carrello
    if not carrello:
        print("üõí Il carrello √® gi√† vuoto.")
        return

    # Chiede conferma prima di svuotare
    conferma = input("Sei sicuro di voler svuotare il carrello? (s/n): ").strip().lower()
    if conferma == "s":
        carrello.clear()
        print("‚úÖ Carrello svuotato con successo.")
    else:
        print("Operazione annullata.")

    
     
# --- FUNZIONE MENU PRINCIPALE ---


def menu_principale():
    """Visualizza il menu principale dell'applicazione.
    
    Stampa tutte le opzioni disponibili del menu principale
    numerandole da 1 a 8.

    Returns:
        None
    """
    print("\n--- Menu Principale ---")
    print("1. Inserisci un nuovo capo")
    print("2. Visualizza tutti i capi")
    print("3. Modifica un capo")
    print("4. Elimina capi")
    print("5. Cerca un capo")
    print("6. Funzioni avanzate")
    print("7. Visualizza carrello")
    print("8. Esci")

.
def main():
    """Funzione principale che gestisce il ciclo del programma.
    
    Presenta il menu principale e gestisce la selezione delle opzioni
    finch√© l'utente non sceglie di uscire (opzione 8).

    Returns:
        None
    """
    while True:
        menu_principale() 
        print("")
        scelta = input("Scegli un'opzione (1-7): ")
        print("")
        
        if scelta == "1":
            insert_capo()
        elif scelta == "2":
            view_capi()
        elif scelta == "3":
            modifica_capo()
        elif scelta == "4":
            elimina_capo()
        elif scelta == "5":
            ricerca_capo()
        elif scelta == "6":
            funzioni_avanzate()  
        elif scelta == "7":
            menu_carrello()      
        elif scelta == "8":
            print("Arrivederci!")
            break
        else:
            print("Opzione non valida. Riprova.")

# --- FUNZIONE SOTTO MENU AVANZATE ---


import matplotlib.pyplot as plt
def funzioni_avanzate():
    """Sottomenu per le funzionalit√† avanzate del programma.
    
    Offre opzioni per filtrare i capi e visualizzare statistiche.
    Continua fino a quando l'utente sceglie di tornare al menu principale.

    Returns:
        None
    """
    while True:
        print("\n--- Funzioni Avanzate ---\n")
        print("1. Ricerca tramite filtri")
        print("2. Statistiche generali")
        print("3. Torna al men√π principale")
        
        scelta_avanzate = input("\nScegli un'opzione (1-3): ")
        
        if scelta_avanzate == "1":
            filtra_capi()
        elif scelta_avanzate == "2":
            statistiche_capi()
        elif scelta_avanzate == "3":
            break  # Esce dal sottomenu e ritorna al main
        else:
            print("Opzione non valida. Riprova.")

```
